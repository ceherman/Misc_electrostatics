# Hubbuch model

```python
ion_str_vals = np.arange(0.05, 1.11, 0.01)
```

```python
Keq_vals_phq_blg = []
Keq_vals_pxq_blg = []
Keq_vals_phq_ova = []
Keq_vals_pxq_ova = []

for ion_str in ion_str_vals:
    sol = base_classes.solution(7, ion_str)
    
    phq = h.resin(solution=sol, ligand='quat_amine', surf_dens=2.55e-6) 
    pxq = h.resin(solution=sol, ligand='quat_amine', surf_dens=3.72e-6) 

    blg = h.protein(solution=sol, seq=seq_blg, name='blg')
    ova = h.protein(solution=sol, seq=seq_ova, name='ova')
    
    Keq_vals_phq_blg.append(h.get_Keq_nd(blg, phq, phi_poros_blg))
    Keq_vals_pxq_blg.append(h.get_Keq_nd(blg, pxq, phi_poros_blg))
    Keq_vals_phq_ova.append(h.get_Keq_nd(ova, phq, phi_poros_ova))
    Keq_vals_pxq_ova.append(h.get_Keq_nd(blg, pxq, phi_poros_ova))
```

```python
Keq_vals_sep_lys_5 = []
Keq_vals_sep_lys_7 = []
Keq_vals_sep_lys_9 = []

for ion_str in ion_str_vals:
    sol_5 = base_classes.solution(5, ion_str)
    sol_7 = base_classes.solution(7, ion_str)
    sol_9 = base_classes.solution(9, ion_str)

    seph_5 = h.resin(solution=sol_5, ligand='sulfonate', surf_dens=4.73e-6)
    seph_7 = h.resin(solution=sol_7, ligand='sulfonate', surf_dens=4.73e-6)
    seph_9 = h.resin(solution=sol_9, ligand='sulfonate', surf_dens=4.73e-6)

    lys_5 = h.protein(solution=sol_5, seq=seq_lys, name='lys')
    lys_7 = h.protein(solution=sol_7, seq=seq_lys, name='lys')
    lys_9 = h.protein(solution=sol_9, seq=seq_lys, name='lys')
    
    Keq_vals_sep_lys_5.append(h.get_Keq_nd(lys_5, seph_5, phi_seph_lys))
    Keq_vals_sep_lys_7.append(h.get_Keq_nd(lys_7, seph_7, phi_seph_lys))
    Keq_vals_sep_lys_9.append(h.get_Keq_nd(lys_9, seph_9, phi_seph_lys))
```

```python
fig, ax = plt.subplots(1, 3, sharex=True, sharey=True)
fig.set_size_inches(20, 6.5)
ax[0].set_ylabel('$K_{eq}$', rotation=0, labelpad=30)
plt.subplots_adjust(wspace=0.05)
for i, title in enumerate([r'$\beta$-lactoglobulin at pH 7', 'Ovalbumin at pH 7', 'Lysozyme on SP Sepharose FF']):
    ax[i].tick_params(axis='both', which='major', length=10, width=1, direction='in', top=True, bottom=True, left=True, right=True)
    ax[i].tick_params(axis='both', which='minor', length=5, width=1, direction='in', top=True, bottom=True, left=True, right=True)
    ax[i].set_xscale('log')
    ax[i].set_yscale('log')
    ax[i].set_xlabel('Ionic strength [mM]')
    ax[i].set_title(title)

# First two panels
for i, protein in enumerate(aex_proteins):
    for resin in resins_exp:
        df_series = df_exp_aex[(df_exp_aex['resin'] == resin) & (df_exp_aex['protein'] == protein)]
        ax[i].scatter(df_series['ion_str_mM'], df_series['keq'], marker=markers_aex[resin], facecolors='none', edgecolors=colors_aex[resin], label=resin_name[resin])
        
        if resin == 'phq' and protein == 'blg':
            model_Keq = Keq_vals_phq_blg
        elif resin == 'phq' and protein == 'ova':
            model_Keq = Keq_vals_phq_ova
        elif resin == 'pxq' and protein == 'blg':
            model_Keq = Keq_vals_pxq_blg
        elif resin == 'pxq' and protein == 'ova':
            model_Keq = Keq_vals_pxq_ova
        
        if resin == 'caq':
            pass
        else:
#             ax[i].plot(ion_str_vals*1000, model_Keq, marker=markers_aex[resin], color=colors_aex[resin])
            ax[i].plot(ion_str_vals*1000, model_Keq, color=colors_aex[resin])
            
# Last panel
for pH in cex_pH:
    df_series = df_exp_cex[df_exp_cex['pH'] == pH]
    ax[2].scatter(df_series['ion_str_mM'], df_series['keq'], marker=markers_cex[pH], facecolors='none', edgecolors=colors_cex[pH], label=f'pH {pH}')
    
    if pH == 5:
        model_Keq = Keq_vals_sep_lys_5
    elif pH == 7:
        model_Keq = Keq_vals_sep_lys_7
    elif pH == 9:
        model_Keq = Keq_vals_sep_lys_9
        
#     ax[2].plot(ion_str_vals*1000, model_Keq, marker=markers_cex[pH], color=colors_cex[pH])
    ax[2].plot(ion_str_vals*1000, model_Keq, color=colors_cex[pH])
    
    
xlim = ax[0].set_xlim(50, 1100)
ylim = ax[0].set_ylim(0.07, 2e3) 
for i in range(3):
    ax[i].legend(handlelength=0.3, loc='best', frameon=True)
    
fig.savefig(f'./images/hubbuch_z0_2A.pdf', bbox_inches='tight')
fig.savefig(f'./images/hubbuch_z0_2A.png', dpi=300, bbox_inches='tight')
```
