# Load and plot data

```python
folder = './data/problem_demo/'

files = os.listdir(folder)
files.sort()

nacl_vals = {}
dfs = {}
for f in files:
    nacl_vals[f] = float(f[8:11])
    dfs[f] = akta.read_data_normalized(folder + f)
```

```python
vol_c = 0.8246680715673207 # [ml]

for f in files:
    dfs[f]['cv'] = dfs[f].volume_ml/vol_c
```

```python
fig, ax1 = my_plot.instantiate_fig(10, 7, xlabel='CV', ylabel='A215 [mAU]')
ax2 = my_plot.get_twinx(ax1, ylabel='Conductivity [mS/cm]')

for f in files:
    ax1.plot(dfs[f].cv, dfs[f].uv_215_mAU, label=f'{nacl_vals[f]:.0f} mM')
    ax2.plot(dfs[f].cv, dfs[f].cond_mS_cm, ':')

ax2.set_ylim(-5, 180)
ax1.axhline(color='black')

# ax1.set_xlim(54, 57)

my_plot.set_layout(fig, ax1, legend=True)
# fig.savefig('./images/problem_demo_data.png', dpi=300)
```

# Estimate fraction of chromatogram area in the isocratic elution and wash

```python
cut_pts =\
{'cat phq 200.xls':(0, 13, 18),
 'cat phq 150.xls':(0, 13, 18),
 'cat phq 100.xls':(0, 25, 30),
 'cat phq 075.xls':(0, 49, 53),
 'cat phq 050.xls':(0, 54, 57)}
```

```python
results = []

for f in files:
    df = dfs[f]
    nacl = nacl_vals[f]
    lo, mid, hi = cut_pts[f]
    
    df_entire = df[df.cv < hi]
    df_isocratic = df[df.cv < mid]
    df_wash = df[(mid < df.cv) & (df.cv < hi)]
    
    entire = integrate.simps(df_entire.uv_215_mAU, df_entire.cv)
    isocratic = integrate.simps(df_isocratic.uv_215_mAU, df_isocratic.cv)
    wash = integrate.simps(df_wash.uv_215_mAU, df_wash.cv)
    assert abs(entire - isocratic - wash) < 0.1
    
    f_iso  = isocratic/entire
    f_wash = 1.0 - f_iso
    results.append([nacl, f_iso, f_wash])
    
df_frac = pd.DataFrame(results, columns=['nacl', 'f_iso', 'f_wash'])
df_frac
```

```python
fig, ax = my_plot.instantiate_fig(xlabel='Ionic strength [mM]', ylabel='Wash fraction')
ax.scatter(df_frac.nacl+41.0, df_frac.f_wash)
ax.set_ylim(0, 1)
ax.set_xscale('log')
ax.set_xlim(10, 1000)
my_plot.set_layout(fig, ax)
fig.savefig('./images/problem_demo_wash_fraction.png', dpi=300)
```
