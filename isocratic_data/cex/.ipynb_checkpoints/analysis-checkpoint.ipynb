# First cleaning of data

The following occurs when the (split) raw data files are loaded:
- Time is converted to volume and column volume
- The first absorbance measurement is used to zero the profile
- A linear baseline drift correction is applied
- The first moment is estimated (and the extra-column volume is subtracted)

```python
folder = './data/split_files/'
files  = os.listdir(folder)
files.sort()

data_files = {}
for f in files:
    pH = float(f[3])
    if pH == 5:
        buffer_ion_str = 52
    elif pH == 7:
        buffer_ion_str = 52
    elif pH == 9:
        buffer_ion_str = 20
    else:
        raise ValueError()
        
    data_files[f] = my_fun.run_data(folder, f, buffer_ion_str)
```

# Inspect the data
## Create reference maps

```python
## For reference 
pH_vals = [5, 7, 9]

# Enumeration maps
pH_number = {}
for i, pH in enumerate(pH_vals):
    pH_number[pH] = i

# Map of NaCl concentrations used at each pH
nacl_used = {}
for pH in pH_vals:
    nacl_used[pH] = []
for f in files:
    data = data_files[f]
    current_list = nacl_used[data.pH]
    if data.nacl not in current_list:
        nacl_used[data.pH].append(data.nacl)
        nacl_used[data.pH].sort()
```

# $k'$

```python
for f in files:
    data = data_files[f]
    data.eps_t = 0.897
    data.phi = (1.0 - data.eps_t)/data.eps_t
    data.get_kprime()
    data.get_keq()
```

```python
# Color and symbol maps
colors =\
{5:'black',
 7:'red',
 9:'blue'
}

markers =\
{5:'o',
 7:'^',
 9:'s'
}

# Function to not show replicate legend entries
def legend_without_duplicate_labels(ax, loc='best', handlelength=0.5):
    handles, labels = ax.get_legend_handles_labels()
    unique = [(h, l) for i, (h, l) in enumerate(zip(handles, labels)) if l not in labels[:i]]
    ax.legend(*zip(*unique), loc=loc, handlelength=handlelength, frameon=False)
```

```python
# Specify figure dimensions
label_size = 20
tick_label_size = 16

fig, ax = my_plot.instantiate_fig()
ax.set_xlabel('Ionic strength [mM]', fontsize=label_size)
ax.set_ylabel('$k\'$', fontsize=label_size, rotation=0, labelpad=label_size)
ax.tick_params(axis='both', which='major', length=8, width=1, direction='in')
ax.tick_params(axis='both', which='minor', length=4, width=1, direction='in')

# Add the data
for f in files:
    data = data_files[f]
    if data.nacl < 950:
        color  = colors[data.pH]
        marker = markers[data.pH]
        ax.scatter(data.ion_str, data.kprime, marker=marker, facecolors='none', 
                    edgecolors=color, label=f'pH {data.pH}')
        
# Add the legend
legend_without_duplicate_labels(ax)

# Set axis scales
ax.set_ylim(0.01, 1000)
ax.set_xlim(10, 1000)
ax.set_xscale('log')
ax.set_yscale('log')

# Format and save the figure
my_plot.set_layout(fig)
# fig.savefig('./images/kprime.png', dpi=300)
```

```python
results = []

for f in files:
    data = data_files[f]
    if data.nacl < 950:
        results.append([data.resin, data.protein, data.nacl, data.ion_str, data.pH, data.replicate, data.cv_moment, data.eps_t, data.phi, data.kprime, data.keq])
        
df_results = pd.DataFrame(results, columns=['resin', 'protein', 'nacl_mM', 'ion_str_mM', 'pH', 'replicate', 'cv_moment', 'eps_t', 'phi', 'kprime', 'keq'])
df_results.sort_values(by=['resin', 'protein', 'pH', 'nacl_mM', 'replicate'], inplace=True, ignore_index=True)
```
